**Data Cleaning**

```python
#drop unncecessary from domain knowledge
df.drop(columns= ['date','view','sqft_above','sqft_basement','zipcode','lat','long','sqft_living15','sqft_lot15'],inplace = True)
df.head()
```

```python
df['total_sqft'] = df['sqft_living'] + df['sqft_lot']
df['age'] = 2024 - df['yr_built']
df.drop(columns= ['yr_built','sqft_living','sqft_lot'],inplace = True)
df.head()
```

```python
df_yr_renovated = df['yr_renovated'].dropna(inplace= False)
```
