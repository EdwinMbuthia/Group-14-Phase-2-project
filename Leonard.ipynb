# Feature Selection

```python
# select most important features
df_transformed = df_transformed[['grade','bathrooms','bedrooms','floors','price']]
df_transformed.head()
```

## Data Cleaning

Checking for outliers

```python
import matplotlib.pyplot as plt
# Plotting box plots for each column
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_transformed)
plt.title('Box plots of Selected Columns')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

## Get the Outliers 

```python
# Calculate IQR for selected columns
Q1 = df_transformed.quantile(0.25)
Q3 = df_transformed.quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = ((df_transformed < lower_bound) | (df_transformed > upper_bound)).any(axis=1)

# Print indices of outliers
outlier_indices = df_transformed.index[outliers]
```

## Remove the outliers 

```python
# droping all columns with outliers 
df_transformed = df_transformed.drop(outlier_indices)
```

### Test Train Split


```python
#imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Modeling

```python
model = LinearRegression()
```

```python
# training the model
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
df_model = pd.DataFrame({"true":y_test,"pred":y_pred})
df_model.head()
```

## Validation 

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2  = r2_score(y_test,y_pred)
```

```python
print("mse",mse)
print("mae",mae)
print("R2" ,r2)
```
